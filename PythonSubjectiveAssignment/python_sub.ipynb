Q14. Write thePython program to find the **second** most repeated word in a given string?



```python
def RepCount(Str):
    
    #getting set of string as we dont want to check same letters again
    x = set(Str)
    
    #blank dictionary for storing the letter and its count value
    D= {}
    
    #populating the dictionary
    for i in x : 
        n =Str.count(i)
        if (n>1):
            D[i] = n
            
    #sorting the dictionary into a tuple form
    s = [(k, D[k]) for k in sorted(D, key=D.get, reverse=True)] 
    
    #returning second from top element 
    return s[1]


RepCount("hellloww")        

```
